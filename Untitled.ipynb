```python
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
from scipy.optimize import minimize
import scipy.filter as sp
```

```python
import numpy as np 
from scipy.signal import butter,filtfilt
```

#  Step 1: Generate a  signal

```python

np.random.seed(0)
time = np.linspace(0, 1,126 )  
base_signal = np.sin(2 * np.pi * 5 * time)  
noise = np.random.normal(0, 0.5, time.shape)  
noisy_signal = base_signal + noise
```

# step 2 : design the low pass filter 

```python

b, a = butter(3,0.04, btype='low')
```

# step 3 apply the filter 

```python
filteredsignal=filtfilt(b,a,noisy_signal)
```

# SHOWING THE GRAPHS

```python
plt.plot(time,noisy_signal)
plt.grid()
plt.xlabel('Time ')
plt.ylabel('Noisy Signal ')
plt.title('Graph of  Noisy signal vs Time ')
```

```python
# plotting the filtered signal 
plt.plot(time,filteredsignal)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Filtered Signal ')
plt.title('Graph of Filtered Signal vs Time')
```
